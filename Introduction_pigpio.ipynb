# Installation von Jupyter Notebook auf dem Raspberry Pi

Damit Sie den Code eines Notebooks direkt auf dem Raspberry Pi ausführen können, müssen Sie *Jupyter Notebook* installieren.

Dafür können Sie folgendes Installations-Skript herunterladen und in ihrem Home-Verzeichnis (`cd ~`) des Raspberry Pis ausführen.

**Hinweis: Ihr Raspberry Pi benötigt eine Internetverbindung.**

[Zum Script](./setup_jupyter_py.sh)

Ausführen des Skripts:

```python
chmod +x setup_jupyter_py.sh
./setup_jupyter_py.sh
```

Nachdem die Installation erfolgreich war, können Sie den Notebook Server starten, indem Sie folgenden Befehl in der Shell ausführen:

```python
jupyter notebook
```

Um sich mit dem Server zu verbinden, können Sie entweder direkt auf dem Pi einen Webbrowser öffnen und in die Adresszeile `http://localhost:7777/lab` eingeben, oder von einem Rechner im selben Netzwerk über `http://<lokale-ip-des-pi>:7777/lab` darauf zugreifen.
