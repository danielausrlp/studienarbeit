# Einführung in den SoS-Kernel (Script of Scripts)

Dieses Notebook gibt einen Überblick über den SoS-Kernel für Jupyter, der die Ausführung von Multi-Language-Skripten innerhalb eines Notebooks ermöglicht.

## Installation

Es wird ein Installationsskript bereitgestellt, was den *jupyter notebook* Server direkt mit dem SoS-Kernel und C++-Kernel in einem *Docker-Container* installiert.

[jupyter-sos_install.sh](jupyter-sos_install.sh)

```python
!chmod +x jupyter-sos_install.sh
!./jupyter-sos_install.sh
```

## Ausführung

Den Docker-Container starten Sie im Terminal mit:

```python
!sos_start
```

und stoppen ihn mit:

```python
!sos_stop
```

Sobald der Container läuft, können Sie über einen Webbrowser auf den Jupyter-Server zugreifen:

-   Lokaler Zugriff: http://localhost:8888

-   Zugriff über einen Server: http://(ip-adresse):8888

Dort lassen sich Notebooks erstellen, bearbeiten und ausführen.

Während der Bearbeitung eines Notebooks können Sie über das Drop-Down-Menü den Kernel (also die Programmiersprache) für jede Code-Zelle auswählen.

Die erstellten Notebooks werden standardmäßig im Home-Verzeichnis des Nutzers unter:

```python
~/sosNotebooks
```

gespeichert.

Hinweis: Standardmäßig sind bereits ein Python- und ein C++-Kernel verfügbar. Zusätzliche Kernels können bei Bedarf installiert werden.
