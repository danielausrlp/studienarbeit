# Intro

In diesem Notebook vergleichen wir die Ausführungsgeschwindigkeit und Effizienz von C++ und Python anhand ausgewählter Algorithmen. Ziel ist es, typische Szenarien zu analysieren, in denen die Wahl der Programmiersprache einen messbaren Einfluss auf die Laufzeit hat.

# Unterschiede C++ vs. Python

C++ ist eine kompilierte Sprache und kann durch Optimierungen (z. B. Compiler-Flags) sehr nahe an der Hardware arbeiten. Python hingegen ist interpretiert und dynamisch typisiert, was grundsätzlich Overhead erzeugt. Daher kann man erwarten, dass C++ grundsätzlich schneller Algorithmen ausführt als Python. Jedoch wird Python dank der einfacheren und schnelleren Entwicklung gerne für "Prototypen" genutzt, und die rechenintensiven Teile des Programms später in C++ umgesetzt.

# Erstes Beispiel: Rekursive Fibonacci Funktion

Als erstes benchmarken wir die rekursive Fibonacci Funktion (siehe https://de.wikipedia.org/wiki/Fibonacci-Folge und https://www.geeksforgeeks.org/dsa/program-for-nth-fibonacci-number/).

**C++:**

Hiermit erstellen wir die Funktion fib_cpp(n), die uns die n'te Fibonacci-Nummer errechnet:

Jetzt wollen wir die Zeit messen, die die Funktion zum Ausführen braucht, um die 20'te Fibonacci-Nummer zu errechnen. Dazu machen wir Gebrauch der *chrono*-Bibliothek, die uns eine *high_resolution_clock* bereitstellt. Diese gibt uns die genaue momentane Zeit zurück.

```python
%%writefile fib.cpp
//Programmcode in Datei fib.cpp speichern, um es später zu kompilieren

#include <iostream>
#include <chrono>

int fib_cpp(int n) {

    if(n < 2)
        return n;

    return fib_cpp(n-1) + fib_cpp(n-2);

}

int main() {

  //Hier nutzen wir die chrono Bibliothek, um die momentane Zeit zu speichern
  auto start = std::chrono::high_resolution_clock::now();

  int result = fib_cpp(20);

  //Zeit nach dem Funktionsaufruf speichern
  auto end = std::chrono::high_resolution_clock::now();

  //Die Differenz von Endzeit und Startzeit liefert uns die Ausführungszeit
  std::chrono::duration<double> time_cpp = end - start;

  std::cout << "fib_cpp(20) took " << std::fixed << time_cpp.count() << " seconds to finish. Result: " << result << std::endl;

  return 0;

}

```

Um das C++-Programm auf Colab auszuführen, müssen wir es mit *g++* kompilieren. Hiermit geben wir dann die Ausführungszeit und das Ergebnis aus:

```python
!g++ fib.cpp -o fib
!./fib
```

**Python:**

Nun werden wir in Python die selbe Funktion erstellen:

```python
import time

def fib_py(n):
    if n < 2:
        return n
    return fib_py(n-1) + fib_py(n-2)
```

Wir verwenden das selbe Prinzip, wie im C++ Beispiel, um die Ausführungszeit zu messen:

```python
start = time.perf_counter()

result = fib_py(20)

end = time.perf_counter()

time_py = end - start
```

Jetzt können wir die Ausführungszeit und das Ergebnis ausgeben:

```python
print(f"fib_py(20) took {time_py:.6f} seconds to finish. Result: {result}")
```


Man sollte erkennen, dass die Ausführungszeit von Python **höher** ist, als die von C++.

Dieser Unterschied wird noch deutlicher, falls man die Komplexität erhöht. In unserem Beispiel werden wir jetzt die 40'te Fibonacci-Nummer errechnen. Es könnte sein, dass die Berechnung etwas Zeit in Anspruch nimmt.

**C++:**

```python
%%writefile fib.cpp
#include <iostream>
#include <chrono>

int fib_cpp(int n) {

    if(n < 2)
        return n;

    return fib_cpp(n-1) + fib_cpp(n-2);

}

int main() {

  auto start = std::chrono::high_resolution_clock::now();

  int result = fib_cpp(40);

  auto end = std::chrono::high_resolution_clock::now();

  std::chrono::duration<double> time_cpp = end - start;

  std::cout << "fib_cpp(40) took " << std::fixed << time_cpp.count() << " seconds to finish. Result: " << result << std::endl;

  return 0;

}
```

 Wie man sehen kann, ist Python jetzt **erheblich** langsamer als C++.

**C++**:

```python
%%writefile fib.cpp
#include <iostream>
#include <chrono>

#define MAX_ITERATIONS 8

int fib_cpp(int n) {

    if(n < 2)
        return n;

    return fib_cpp(n-1) + fib_cpp(n-2);

}


int main() {


  for (size_t i = 1; i < MAX_ITERATIONS + 1; i++) {

    auto start = std::chrono::high_resolution_clock::now();
    int result = fib_cpp(i * 5);
    auto end = std::chrono::high_resolution_clock::now();

    std::chrono::duration<double>time_cpp = end - start;

    std::cout << std::fixed <<"fib_cpp(" << i * 5 << ") took " << time_cpp.count() << " seconds to finish. Result: " << result << std::endl;

  }

  return 0;

}


```

# Zweites Beispiel: Primzahlen-Generator

Wir benchmarken jetzt eine (primitive, aber fürs Benchmarken geeignet) Primzahl-Generator-Funktion. Sie gibt einen *vector* mit Primzahlen bis n zurück.

**C++:**

Nun werden wir wie im Fibonacci-Beispiel mehrere Parameter zum Benchmarken ausprobieren (Primzahlen von 1000 bis 10000):

```python
%%writefile primes.cpp
#include <vector>
#include <iostream>
#include <chrono>

#define MAX_ITERATIONS_PRIMES 10
#define PRIMES_STEP_COUNT 1000

std::vector<int> generatePrimes_cpp(int n) {

    std::vector<int> primes;

    if (n <= 1)
        return primes;

    if (n < 3) {
        primes.push_back(2);
        return primes;
    }

    primes.push_back(2);
    primes.push_back(3);

    for (size_t i = 5; i <= n; i += 2) {

        bool shouldAdd = true;

        for (size_t j = 3; j * j <=i; j += 2) {

            if (i % j == 0) {
                shouldAdd = false;
                break;
            }

        }

        if (shouldAdd)
            primes.push_back(i);

    }

    return primes;

}


int main() {

  for (size_t i = 1; i < MAX_ITERATIONS_PRIMES; i++) {

    std::vector<int> primes;

    auto start = std::chrono::high_resolution_clock::now();

    primes = generatePrimes_cpp(i*PRIMES_STEP_COUNT);

    auto end = std::chrono::high_resolution_clock::now();

    std::chrono::duration<double> elapsed = end - start;

    std::cout << "Generated " << std::fixed << primes.size() << " primes in " << elapsed.count() << " seconds." << std::endl;

    }


  return 0;

}
```

```python
!g++ primes.cpp -o primes
!./primes
```


**Python:**

```python
import time

def generatePrimes_py(n):
    primes = []

    if n <= 1:
        return primes

    if n < 3:
        primes.append(2)
        return primes

    primes.append(2)
    primes.append(3)

    for i in range(5, n+1, 2):

        shouldAdd = True

        for j in range(3, int(i**0.5) + 1,2):

            if i % j == 0:
                shouldAdd = False;
                break

        if shouldAdd:
            primes.append(i)

    return primes
```

Zeiten ausgeben:

```python
MAX_ITERATIONS_PRIMES = 10
PRIMES_STEP_COUNT = 1000

for i in range(1, MAX_ITERATIONS_PRIMES, 1):

    start = time.perf_counter()

    primes = generatePrimes_py(i*PRIMES_STEP_COUNT)

    end = time.perf_counter()

    elapsed = end - start

    print(f"Generated {len(primes)} primes in {elapsed:.6f} seconds.")
```

Auch hier kann man erkennen, dass Python **deutlich länger** braucht, als C++.
